```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('bank.csv')

# Display the first few rows of the dataset
print(df.head())

#  Bar plot of job distribution
job_counts = df['job'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(job_counts.index, job_counts.values)
plt.title('Distribution of Jobs')
plt.xlabel('Job')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python

# Pie chart of marital status distribution
marital_counts = df['marital'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(marital_counts.values, labels=marital_counts.index, autopct='%1.1f%%')
plt.title('Marital Status Distribution')
plt.show()

```

```python
# Histogram of age distribution
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=20, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python

#  Bar plot of education level
education_counts = df['education'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(education_counts.index, education_counts.values)
plt.title('Education Level Distribution')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python

# Line plot of campaign duration
plt.figure(figsize=(10, 6))
plt.plot(df['duration'])
plt.title('Campaign Duration')
plt.xlabel('Index')
plt.ylabel('Duration')
plt.show()

```

```python
# Scatter plot of campaign duration vs. previous contact
plt.figure(figsize=(10, 6))
plt.scatter(df['duration'], df['previous'])
plt.title('Campaign Duration vs. Previous Contact')
plt.xlabel('Duration')
plt.ylabel('Previous Contact')
plt.show()


```

```python
# Box plot of campaign duration by marital status
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['marital'] == 'married']['duration'], df[df['marital'] == 'single']['duration'],
             df[df['marital'] == 'divorced']['duration']])
plt.title('Campaign Duration by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Duration')
plt.xticks([1, 2, 3], ['Married', 'Single', 'Divorced'])
plt.show()
```

```python
# Create a figure and subplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(16, 10))

# Visualization 1: Bar plot of job distribution
job_counts = df['job'].value_counts()
axes[0, 0].bar(job_counts.index, job_counts.values)
axes[0, 0].set_title('Distribution of Jobs')

# Visualization 2: Pie chart of marital status distribution
marital_counts = df['marital'].value_counts()
axes[0, 1].pie(marital_counts.values, labels=marital_counts.index, autopct='%1.1f%%')
axes[0, 1].set_title('Marital Status Distribution')

# Visualization 3: Histogram of age distribution
axes[0, 2].hist(df['age'], bins=20, edgecolor='black')
axes[0, 2].set_title('Age Distribution')

# Visualization 4: Bar plot of education level
education_counts = df['education'].value_counts()
axes[0, 3].bar(education_counts.index, education_counts.values)
axes[0, 3].set_title('Education Level Distribution')

# Visualization 5: Line plot of campaign duration
axes[1, 0].plot(df['duration'])
axes[1, 0].set_title('Campaign Duration')

# Visualization 6: Scatter plot of campaign duration vs. previous contact
axes[1, 1].scatter(df['duration'], df['previous'])
axes[1, 1].set_title('Campaign Duration vs. Previous Contact')

# Visualization 7: Box plot of campaign duration by marital status
axes[1, 2].boxplot([df[df['marital'] == 'married']['duration'], df[df['marital'] == 'single']['duration'],
             df[df['marital'] == 'divorced']['duration']])
axes[1, 2].set_title('Campaign Duration by Marital Status')
axes[1, 2].set_xticklabels(['Married', 'Single', 'Divorced'])

# Remove empty subplot
axes[1, 3].remove()

# Adjust spacing and layout
plt.tight_layout()

# Show the figure
plt.show()
```
